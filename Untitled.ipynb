```python
def list_comp(df, a, b, new_column_name):
    df[new_column_name] = df.apply(lambda row: 0 if row[a] >0 or row[b] > 0 else 1, axis=1)
    return df
```

```python
def create_pattern_column(df, column_name, patterns, new_column_name='Pattern'):
    """ 
    Create a new column based on patterns in a string column.

    Parameters:
    df (DataFrame): Input DataFrame.
    column_name (str): Name of the column containing strings to search.
    patterns (list of tuples): List of tuples where each tuple contains a pattern (regular expression) and its corresponding label.
    new_column_name (str, optional): Name of the new column to be created. Default is Pattern.
    Returns:
    DataFrame: DataFrame with the new pattern column added.
    """
    # Define a function to check if any pattern is present in the text
    def check_patterns(text):
        if isinstance(text, str):
            for pattern, value in patterns:
                if re.search(pattern, text):
                    return value
            return 0
        else:   
            return 0

    # Apply the function to each row of the specified column
    df[new_column_name] = df[column_name].astype(str).apply(check_patterns)

    return df
        
```

```python
cabin_floor = [
    (r'A', 1),
    (r'B', 2),
    (r'C', 3),
    (r'D', 4),
    (r'E', 5),
    (r'F', 6),
    (r'G', 7),
    (r'N', 0)
]

create_pattern_column(train_df, 'Cabin', cabin_floor, 'Floor')
train_df
    
```

```python
def extract_title(name):
    match = re.search(r'\b(\w+\.)', name)
    return match.group(1) if match else None
```

```python
train_df['Title'] = train_df['Name'].apply(extract_title)
train_df['Title'].value_counts()
```

```python
mean_ages = train_df.groupby('Title')['Age'].mean()
mean_ages
```

```python
 def fill_missing_age(row):
    if pd.isnull(row['Age']):
        return mean_ages[row['Title']]
    return row['Age']
```

```python
train_df['Age'] = train_df.apply(fill_missing_age, axis=1)
train_df['Age'].value_counts()
```
